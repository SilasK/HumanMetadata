```python
curatedMetagenomicData = curatedMetagenomicData.fillna("none")
```

```python
df_filtered = curatedMetagenomicData[curatedMetagenomicData['NCBI_accession'] != 'none']
```

```python
def splitDataFrameList(df,target_column,separator):
    row_accumulator = []

    def splitListToRows(row, separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)

    df.apply(splitListToRows, axis=1, args = (separator, ))
    new_df = pd.DataFrame(row_accumulator)
    return new_df
```

```python
df_filtered_split = splitDataFrameList(df_filtered,'NCBI_accession',';')
```
